```python
#Para el array completo normalizamos
normalizada=cv2.normalize(imagen_array,0, 255, cv2.NORM_MINMAX, dtype=cv2.CV_32F)
```

```python
normalizada.shape
```

```python
#Una componente
plt.imshow(normalizada[:,:,1])
plt.show()
```

```python
import colorsys
R=normalizada[:,:,2]
G=normalizada[:,:,1]
B=normalizada[:,:,0]

YIQ=colorsys.rgb_to_yiq(R,G,B)
```
