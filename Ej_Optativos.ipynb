```python
import numpy as np  # procesamiento matricial
import imageio_ffmpeg # para leer/guardar videos
import matplotlib.pyplot as plt  # para mostrar imagenes
import matplotlib.colors as pltc
%matplotlib inline
import cv2 
```

```python
# Funcion para extraer frames del video 
def FrameCapture(path): 
      
    # path al video
    vidObj = cv2.VideoCapture(path) 
  
    # variable contadora 
    count = 0
  
    # controla los frames extraidos
    success = 1
  
    while success: 
  
        # vidObj object llama a leer 
        # la funcion extrae frames 
        success, image = vidObj.read() 
  
        # guarda los frams con frames 
        cv2.imwrite("frame%d.jpg" % count, image) 
  
        count += 1
  
```

```python
FrameCapture("video.mp4") #Generamos nuestra base de datos de imagenes 
```

A continuacion se leeran distintas imagenes, tanto del data set generado como de otras imagenes elegidas mas coloridas, para luego mostrarlas en 4 distintas formas. Se puede observar que las imgenes del dataset de cerebro no genera uchas diferencias, en comparacion con las mas coloridas. 

```python
img_0 = cv2.imread('frame200.jpg') #seleccionamos algun frame entre el frame0 y frame271.
img_0_1= cv2.cvtColor(img_0, cv2.IMREAD_COLOR) 
img_0_2= cv2.cvtColor(img_0, cv2.IMREAD_GRAYSCALE)
img_0_3= cv2.cvtColor(img_0, cv2.COLOR_BGR2HSV)
img_0_4= cv2.cvtColor(img_0, cv2.COLOR_BGR2YCrCb)
```

```python
imagenes_0=[img_0_1, img_0_2, img_0_3, img_0_4]
titles=['Imagen a color', 'Imagen en grises', 'Imagen en hsv', 'Imagen en YCrCb', 'Imagen en YUV_YV12']
```

```python
plt.figure (figsize=([20, 20]))
for i in range(4):
    plt.subplot(1, 4, i+1), plt.imshow(imagenes_0[i])
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show
```
